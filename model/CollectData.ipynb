```python
import os 
import pandas as pd
import cv2
from cvzone.HandTrackingModule import HandDetector
import numpy as np
import math
import time
from tensorflow import keras
```

```python
def load_images_from_directory(root_dir,image_size=(28,28)):
    images=[]
    labels=[]
    dem=1
    classes=sorted(os.listdir(root_dir))

    for class_index,class_name in enumerate(classes):
        class_dir=root_dir+'/'+class_name

        for filename in os.listdir(class_dir):
            file_path=class_dir+'/'+filename
            print(file_path)
            image=cv2.imread(file_path,cv2.IMREAD_GRAYSCALE).astype(np.uint8)
            image=cv2.resize(image,image_size)
            # image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
            images.append(image)
            labels.append(class_index)

    return np.array(images),np.array(labels),classes
```

```python
data_num_dir='asl_dataset'
X_train,y_train,classes=load_images_from_directory(data_num_dir,(100,100))
```

```python
shuffle_indices = np.random.permutation(X_train.shape[0])
X_train = X_train[shuffle_indices]
y_train = y_train[shuffle_indices]
```
